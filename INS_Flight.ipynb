```python
URL=list()
for src in locations:
    for destin in locations:
        for doj in Date:
            if src!=destin:
                try:    
                    dr=webdriver.Chrome(r"C:\web driver\New Folder\chromedriver_vers_106.exe")
                    dr.get("https://www.easemytrip.com/")
                    dr.maximize_window()
                    time.sleep(5)

                    # Source

                    WebDriverWait(dr,5).until(EC.visibility_of_element_located((By.XPATH,("/html/body/form/div[5]/div[3]/div/div[2]/div/div[1]/div[1]/input[2]"))))
                    click_tag=dr.find_element("xpath","/html/body/form/div[5]/div[3]/div/div[2]/div/div[1]/div[1]/input[2]")
                    click_tag.click()

                    WebDriverWait(dr,5).until(EC.visibility_of_element_located((By.XPATH,("//div[@class='searcityCol']/input"))))
                    searchFrom_tag=dr.find_element("xpath","/html/body/form/div[5]/div[3]/div/div[2]/div/div[1]/div[2]/div[1]/input")
                    searchFrom_tag.send_keys(src)          

                    try:                          
                        WebDriverWait(dr,5).until(EC.presence_of_element_located((By.XPATH,("//ul[@class='ausuggest']/li[1]/div/div/p/span[1]"))))
                        from_tag=dr.find_element("xpath","/html/body/form/div[5]/div[3]/div/div[2]/div/div[1]/div[2]/div[2]/ul/li[1]/div")
                        from_tag.click()                 
                    except:
                        WebDriverWait(dr,5).until(EC.visibility_of_element_located((By.XPATH,("//ul[@class='ausuggest']/li[1]/div/div/p/span[1]"))))
                        from_tag=dr.find_element("xpath","//ul[@class='ausuggest']/li[1]/div/div/p/span[1]")
                        from_tag.click() 

                    #Destination
                    WebDriverWait(dr,5).until(EC.visibility_of_element_located((By.XPATH,("/html/body/form/div[5]/div[3]/div/div[2]/div/div[3]/div[1]/input[2]"))))
                    click_tag=dr.find_element("xpath","/html/body/form/div[5]/div[3]/div/div[2]/div/div[3]/div[1]/input[2]")
                    click_tag.click()

                    WebDriverWait(dr,10).until(EC.visibility_of_element_located((By.XPATH,("/html/body/form/div[5]/div[3]/div/div[2]/div/div[3]/div[2]/div[1]/input"))))
                    searchFrom_tag=dr.find_element("xpath","/html/body/form/div[5]/div[3]/div/div[2]/div/div[3]/div[2]/div[1]/input")
                    searchFrom_tag.send_keys(destin)

                    WebDriverWait(dr,10).until(EC.visibility_of_element_located((By.XPATH,("/html/body/form/div[5]/div[3]/div/div[2]/div/div[3]/div[2]/div[2]"))))
                    from_tag=dr.find_element("xpath","/html/body/form/div[5]/div[3]/div/div[2]/div/div[3]/div[2]/div[2]/ul/li[1]/div")
                    from_tag.click()  

                    # Date

                    WebDriverWait(dr,5).until(EC.presence_of_element_located((By.XPATH,("/html/body/form/div[5]/div[3]/div/div[2]/div/div[4]/div[1]/input"))))
                    click_tag=dr.find_element("xpath","/html/body/form/div[5]/div[3]/div/div[2]/div/div[4]/div[1]/input")
                    click_tag.click()
                    time.sleep(5)
                    for i in range(7):
                        try:

                            WebDriverWait(dr,3).until(EC.visibility_of_element_located((By.ID,(doj))))

                            select_date=dr.find_element(By.ID, doj)
                            select_date.click()
                            break
                        except:
                            next_month=dr.find_element(By.ID,"img2Nex")
                            next_month.click()

                    search_button=dr.find_element("xpath","//div[@class='fss_flex search_colm']")
                    search_button.click()
                    # Url
                    URL.append(dr.current_url)


                    dr.close()
                except:
                    dr.close()
                    continue
              
```

```python
Dict1=dict()
Dict1['Airline']=list()
Dict1['Flight_number']=list()
Dict1['Dep_time']=list()
Dict1['Source']=list()
Dict1['Class']=list()
Dict1['Route']=list()
Dict1['Duration']=list()
Dict1['Stop']=list()
Dict1['Arrival_time']=list()
Dict1['Destination']=list()
Dict1['Fare']=list()
Dict1['Dep_Date']=list()
```

```python
for url in URL[:50]:
    try:  

        dr1=webdriver.Chrome(r"C:\web driver\New Folder\chromedriver_vers_106.exe")
        dr1.get(url)
        dr1.maximize_window()
        time.sleep(5)
        
        airline =dr1.find_elements('xpath',"//div[@class='col-md-7 col-sm-7 padd-lft airl-txt-n']/span[1]")

        number=dr1.find_elements('xpath', "//div[@class='col-md-7 col-sm-7 padd-lft airl-txt-n']/span[2]")

        dep_time =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[1]")

        sorce =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[2]")

        duration=dr1.find_elements( 'xpath',"//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']/span[1]")

        stop =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']/span[2]")

        arrivaltime=dr1.find_elements('xpath', "//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[1]")

        destination =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[2]")

        fare =dr1.find_elements('xpath',"//div[@class='col-md-8 col-sm-8 col-xs-9 txt-r6-n exPrc ng-binding']")

        for a,n,dep,s,dur,st,at,dest,fr, clk in zip(airline,number,dep_time,sorce,duration,stop,arrivaltime,destination,fare, range(len(airline))):
            try:

                Dict1['Airline'].append(a.text)
                Dict1['Flight_number'].append(n.text)
                Dict1['Dep_time'].append(dep.text)
                Dict1['Source'].append(s.text)
                Dict1['Duration'].append(dur.text)
                Dict1['Stop'].append(st.text)

                Dict1['Arrival_time'].append(at.text)
                Dict1['Destination'].append(dest.text)
                Dict1['Fare'].append(fr.text)

                x=dr1.find_element(by='id',value=str(clk))
                x.click()
                route=dr1.find_elements('xpath',"//span[@class='txt-r9']")
                Dict1['Route'].append([i.text for i in route])
                flightclass =dr1.find_element('xpath',"//span[@class='txt-b3 ng-binding']")
                Dict1['Class'].append(flightclass.text)
                dep_date=dr1.find_element('xpath',"//span[@class='txt-b2nn ng-binding']")
                Dict1['Dep_Date'].append(dep_date.text)
            except:
                continue
        dr1.close()   
                
    except:
        dr1.close()
        pass
            
df1=pd.DataFrame(Dict1)     
df1.to_csv(r"C:\Users\A\Desktop\Flight_URl\flight_prediction_data_01")
                
```
