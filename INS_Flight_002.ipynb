```python
import pandas as pd
import numpy as np
from selenium import webdriver 
import warnings
warnings.filterwarnings("ignore")
import time
import requests

from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import re

import pandas as pd
from datetime import datetime
import random
```

```python
URL=pd.read_csv(r"C:\Users\A\Desktop\Flight_URl\URL_data.csv")
URL
```

```python
Dict1=dict()
Dict1['Airline']=list()
Dict1['Flight_number']=list()
Dict1['Dep_time']=list()
Dict1['Source']=list()
Dict1['Class']=list()
Dict1['Route']=list()
Dict1['Duration']=list()
Dict1['Stop']=list()
Dict1['Arrival_time']=list()
Dict1['Destination']=list()
Dict1['Fare']=list()
Dict1['Dep_Date']=list()
```

```python
n = 200
URL
random.seed(56)
random_urls=random.sample(sorted(URL['0']), n)
print(random_urls)
```

```python
for url in random_urls[:50]:
    try:  

        dr1=webdriver.Chrome(r"C:\web driver\New Folder\chromedriver_vers_106.exe")
        dr1.get(url)
        dr1.maximize_window()
        time.sleep(5)
        
        airline =dr1.find_elements('xpath',"//div[@class='col-md-7 col-sm-7 padd-lft airl-txt-n']/span[1]")

        number=dr1.find_elements('xpath', "//div[@class='col-md-7 col-sm-7 padd-lft airl-txt-n']/span[2]")

        dep_time =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[1]")

        sorce =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[2]")

        duration=dr1.find_elements( 'xpath',"//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']/span[1]")

        stop =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']/span[2]")

        arrivaltime=dr1.find_elements('xpath', "//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[1]")

        destination =dr1.find_elements('xpath',"//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[2]")

        fare =dr1.find_elements('xpath',"//div[@class='col-md-8 col-sm-8 col-xs-9 txt-r6-n exPrc ng-binding']")

        for a,n,dep,s,dur,st,at,dest,fr, clk in zip(airline,number,dep_time,sorce,duration,stop,arrivaltime,destination,fare, range(len(airline))):
            try:

                Dict1['Airline'].append(a.text)
                Dict1['Flight_number'].append(n.text)
                Dict1['Dep_time'].append(dep.text)
                Dict1['Source'].append(s.text)
                Dict1['Duration'].append(dur.text)
                Dict1['Stop'].append(st.text)

                Dict1['Arrival_time'].append(at.text)
                Dict1['Destination'].append(dest.text)
                Dict1['Fare'].append(fr.text)

                x=dr1.find_element(by='id',value=str(clk))
                x.click()
                route=dr1.find_elements('xpath',"//span[@class='txt-r9']")
                Dict1['Route'].append([i.text for i in route])
                flightclass =dr1.find_element('xpath',"//span[@class='txt-b3 ng-binding']")
                Dict1['Class'].append(flightclass.text)
                dep_date=dr1.find_element('xpath',"//span[@class='txt-b2nn ng-binding']")
                Dict1['Dep_Date'].append(dep_date.text)
            except:
                continue
        dr1.close()   
                
    except:
        dr1.close()
        pass
            
df1=pd.DataFrame(Dict1)     
df1.to_csv(r"C:\Users\A\Desktop\Flight_URl\flight_prediction_data_A")
                
```
